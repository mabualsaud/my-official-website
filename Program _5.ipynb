```python
# =============================================================================
# QUESTION 5: INTEGRATED MENU MANAGEMENT SYSTEM
# =============================================================================


# --- PART (i): INITIAL SETUP ---

# Define dictionaries for each course
starters = {"Calamari rings": 6.25, "Chicken Wings": 6.50, "Grilled Halloumi": 5.50}
mains = {"Grilled Salmon": 12.60, "Beef Burger": 11.50, "Vegetable Lasagne": 5.10}
desserts = {"Ice Cream": 3.60, "Lemon Tart": 4.75, "Cheesecake": 5.25}

# Step 1: Create the list called 'menu' as in Q4
menu = [starters, mains, desserts]

# Step 2: Convert the list into a dictionary called 'menu_d'
menu_d = {"Starters": starters, "Mains": mains, "Desserts": desserts}


# Function to print the menu dictionary in an organized and readable format
def print_menu(menu_dict):
    """
    Purpose: Prints the contents of the menu dictionary in an organized format.

    Parameters:
    menu_dict (dict): A dictionary where keys are course names and
                      values are dictionaries of dish names and prices.

    Returns:
    None
    """
    print("--- ORGANIZED RESTAURANT MENU ---")

    # Loop through each course category and its items
    for category, items in menu_dict.items():
        print(f"\n{category}:")

        # Print each dish and its price formatted to 2 decimal places
        for dish, price in items.items():
            print(f"  - {dish}: £{price:.2f}")


# =============================================================================
# --- PART (ii): FUNCTION TO ADD A NEW DISH ---
# =============================================================================

def add_dish(course, name, price):
    """
    Purpose: Adds a new dish entry to a specified course in the menu dictionary.

    Parameters:
    course (str): The course category to add the dish to (e.g., 'Starters').
    name (str): The name of the new dish.
    price (float): The price of the new dish.

    Returns:
    None
    """

    # Check if the course category exists in the menu dictionary
    if course in menu_d:

        # Add the new dish and its price to the appropriate nested dictionary
        menu_d[course][name] = price

        print(f"\nSuccess: '{name}' has been added to {course}.")
        print(f"Updated items in {course}:")

        # Print the updated list of items for this course
        for item, p in menu_d[course].items():
            print(f"  - {item}: £{p:.2f}")

    else:
        # Inform the user if the course name provided is invalid
        print(f"\nError: '{course}' is an invalid course category.")


# =============================================================================
# --- PART (iii): FIND DISH PRICE & MEAL CALCULATION FUNCTIONS ---
# =============================================================================

def find_dish_price(menu_list):
    """
    Purpose: Searches for a dish's price across all categories in the menu list.

    Parameters:
    menu_list (list): A list containing course dictionaries.

    Returns:
    float: The price of the dish if found, otherwise None.
    """

    # Get dish name from user, strip spaces, convert to lowercase for comparison
    item_name = input("Enter dish name: ").strip().lower()

    # Loop through each course dictionary (Outer Loop)
    for course_dict in menu_list:

        # Nested loop through each dish in the course (Inner Loop)
        for dish, price in course_dict.items():

            # Case-insensitive comparison to handle different input formats
            if dish.lower() == item_name:
                return price

    # If not found, print error message and return None
    print("Error: Dish not found in the menu.")
    return None


def calculate_meal_total():
    """
    Purpose: Asks the user to select one dish per course, calculates and
             returns the total meal price. Validates against missing dishes
             and duplicate course selections.

    Parameters:
    None

    Returns:
    float: The total price of the meal if all selections are valid, otherwise None.
    """
    print("\n--- FULL MEAL SELECTION ---")
    print("Please enter one choice for each course (Starter, Main, Dessert):")

    # Ask the user to select one dish per course using find_dish_price from Q4
    print("\n[Selecting Starter]")
    starter_price = find_dish_price(menu)

    print("\n[Selecting Main]")
    main_price = find_dish_price(menu)

    print("\n[Selecting Dessert]")
    dessert_price = find_dish_price(menu)

    # Check if any dish was not found (returned None)
    if starter_price is None or main_price is None or dessert_price is None:
        print("\nInvalid selection: One or more dishes were not found.")
        return None

    # Check if the user selected the same course more than once
    # If all three prices are unique, they likely chose different dishes
    chosen_prices = [starter_price, main_price, dessert_price]
    if len(set(chosen_prices)) != len(chosen_prices):
        print("\nInvalid selection: You cannot select multiple dishes from the same course.")
        return None

    # Calculate the total price for all three courses
    total_price = starter_price + main_price + dessert_price
    print(f"\nSuccess! Your full meal total is: £{total_price:.2f}")

    return total_price


# =============================================================================
# --- PART (iv): EXECUTION ---
# =============================================================================

# 1. Print the full menu in an organized format
print_menu(menu_d)

# 2. Add a new dish to the Starters course
add_dish("Starters", "Spring Rolls", 4.50)

# 3. Run the meal selection and total calculation
final_bill = calculate_meal_total()

# 4. Print the final result
if final_bill is not None:
    print(f"\nFinal Process Complete. Total to pay: £{final_bill:.2f}")
else:
    print("\nFinal Process Complete with errors in selection.")
```

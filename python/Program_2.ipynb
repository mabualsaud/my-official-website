```python
# Part (i): Function to collect 5 unique strings from the user

# Define a function that asks the user for 5 unique strings and stores them in a set
def get_unique_strings():

    """
    Purpose: Asks the user for 5 unique strings using a for loop and stores them in a set.

    Parameters:
    None

    Returns:
    set: A set containing exactly 5 unique strings entered by the user.
    """
    # Initialize an empty set to store unique strings
    user_words = set()

    # Use a for loop with a large range to allow for retries if duplicates are entered
    for i in range(100):
        # Prompt the user for a string
        input_string = input("Please enter a unique string: ")

        # Check if the string is already in the set
        if input_string in user_words:
            print("Duplicate entry, please enter a different string.")
        else:
            # Add the unique string to the set
            user_words.add(input_string)

        # Once we have exactly 5 unique strings, exit the loop
        if len(user_words) == 5:
            break
        # Return the set containing the 5 values
    return user_words


```

```python
# part (ii)
# Define a function that accepts a set of strings and an integer to shorten each string

def shorten_words(words, cutoff):
    """
    Purpose: Returns a set of strings where each string is shortened
             to a maximum length defined by 'cutoff'.

    Parameters:
    words (set of str): Input set of strings
    cutoff (int): Maximum allowed length for each string

    Returns:
    set of str: Set containing shortened strings
    """
    # Create an empty set to store results
    shortened_results = set()

    # Loop through each word in the set
    for word in words:
        # Add the word shortened to the cutoff length
        shortened_results.add(word[:cutoff])

    return shortened_results


```

```python
# part (iii)
# Main code to execute the functions and display results

# 1. Call the first function to get 5 unique words from the user
my_words = get_unique_strings()

# 2. Call the second function using the words collected and a cutoff of 5
final_result = shorten_words(my_words, 5)

# 3. Print the final processed set to the console
print("The final shortened strings are:")
print(final_result)


```

```python
# ==========================================
# Question 2: String Manipulation using Sets
# ==========================================

# Part (i): Define a function to collect 5 unique strings from the user
def get_unique_strings():
    """
    Purpose: Prompts the user to enter 5 unique strings and stores them in a set.

    Parameters:
    None

    Returns:
    set: A set containing exactly 5 unique strings.
    """
    # Initialize an empty set to store unique strings
    user_words = set()

    # Use a for loop with a range to allow retries (ensuring efficiency)
    for i in range(100):
        input_string = input("Please enter a unique string: ")

        # Check if the string is already in the set to avoid duplicates
        if input_string in user_words:
            print("Duplicate entry, please enter a different string.")
        else:
            user_words.add(input_string)

        # Once the set reaches 5 elements, exit the loop efficiently
        if len(user_words) == 5:
            break

    return user_words


# Part (ii): Define a function to shorten strings in a set based on a cutoff
def shorten_words(words, cutoff):
    """
    Purpose: Returns a new set where each string is shortened to the cutoff length.

    Parameters:
    words (set): The input set of strings.
    cutoff (int): The maximum length for each string.

    Returns:
    set: A new set containing the shortened strings.
    """
    # Create an empty set for the processed results
    shortened_results = set()

    # Iterate through each word and apply slicing
    for word in words:
        # Slicing word[:cutoff] is an efficient way to shorten strings
        shortened_results.add(word[:cutoff])

    return shortened_results


# Part (iii): Main execution block to run the functions
# ---------------------------------------------------

# 1. Call the function to get strings from the user
user_input_set = get_unique_strings()

# 2. Call the function to process the set with a cutoff of 5
final_processed_set = shorten_words(user_input_set, 5)

# 3. Print the results to the console for verification
print("\nThe final shortened set of words is:")
print(final_processed_set)
```

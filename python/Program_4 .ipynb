```python
# =============================================================================
# PART (i): INITIALIZING THE MENU DATA STRUCTURE
# =============================================================================

# Define dictionaries for each course
starters = {"Calamari rings": 6.25, "Chicken Wings": 6.50, "Grilled Halloumi": 5.50}
mains = {"Grilled Salmon": 12.60, "Beef Burger": 11.50, "Vegetable Lasagne": 5.10}
desserts = {"Ice Cream": 3.60, "Lemon Tart": 4.75, "Cheesecake": 5.25}

```

```python
# =============================================================================
# PART (ii): CREATING MENU LIST
# ============================================================================
# Create a list called 'menu' containing the dictionaries from part i
menu = [starters, mains, desserts]

```

```python
# =============================================================================
# PART (iii): FUNCTION TO FIND DISH PRICE
# =============================================================================
def find_dish_price(menu_list):
    """
    Purpose: Searches for a dish's price across all categories in the menu list.
    Parameters: menu_list (list) - A list containing dictionaries of courses.
    Returns: price (float) if found, None if not found.
    """

    item_name = input("Enter the name of the dish: ")

    # Loop through each category dictionary (Outer Loop)
    for category in menu_list:

        # Nested loop through items in the category (Inner Loop)
        for item, price in category.items():

            if item.lower() == item_name.lower():
                return price

    # If the dish is not found after searching all categories
    print("Error: Invalid selection. Dish not found.")
    return None

```

```python
# =============================================================================
# PART (iv): RUNNING THE FUNCTION
# =============================================================================

# Call the function and store the returned value
dish_price = find_dish_price(menu)

# Print the result to the console
if dish_price is not None:
    print(f"The price of the selected dish is: {dish_price}")

```

```python
# =============================================================================
# PART (v): FINAL EFFICIENT & DOCUMENTED CODE
# =============================================================================

"""
This program implements a restaurant menu system using dictionaries and a list.
It supports searching for dish prices across multiple categories using nested loops
and handles case-insensitive input to ensure a robust user experience.
"""

# 1. INITIALIZING THE MENU DATA STRUCTURE
# Define dictionaries for each course
starters = {"Calamari rings": 6.25, "Chicken Wings": 6.50, "Grilled Halloumi": 5.50}
mains = {"Grilled Salmon": 12.60, "Beef Burger": 11.50, "Vegetable Lasagne": 5.10}
desserts = {"Ice Cream": 3.60, "Lemon Tart": 4.75, "Cheesecake": 5.25}

# Create a list containing the dictionaries
menu = [starters, mains, desserts]


# 2. FUNCTION TO FIND DISH PRICE
def find_dish_price(menu_list):
    """
    Purpose: Searches for a dish's price across all categories in the menu list.
    Parameters: menu_list (list) - A list containing dictionaries of courses.
    Returns: price (float) if found, None if not found.
    """

    item_name = input("Enter the name of the dish: ")

    # Loop through each category dictionary (Outer Loop)
    for category in menu_list:

        # Nested loop through items in the category (Inner Loop)
        for item, price in category.items():

            # Case-insensitive comparison to ensure "Ice Cream" is found regardless of input case
            if item.lower() == item_name.lower():
                return price

    # If the dish is not found after searching all categories
    print("Error: Invalid selection. Dish not found.")
    return None


# 3. RUNNING THE FUNCTION
# Call the function and store the returned value
dish_price = find_dish_price(menu)

# Print the final result to the console
if dish_price is not None:
    print(f"The price of the selected dish is: {dish_price}")
```

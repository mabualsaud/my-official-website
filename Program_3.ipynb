```python

# Part (i)
# define a function to read text integers from the text file
def read_integers_from_file(filename):
    """
    Purpose: Reads integers from a text file where values are separated by spaces,
             and returns them as a list of integers.

    Parameters:
    filename (str): Name of the file to read from

    Returns:
    list of int: List containing integers from the file
    """

    # Open the input file for reading
    my_file = open(filename, 'r')

    # Read everything
    file_content = my_file.read()

    # Close the file
    my_file.close()

    # Convert text to a list of strings
    string_list = file_content.split()

    # Convert strings to integers
    integers_list = []
    for num in string_list:
        integers_list.append(int(num))

    return integers_list

# =============================================================================
# PART (ii): EXTENDING THE FUNCTION WITH INSERTION SORT (DESCENDING ORDER)
# =============================================================================

# define a function to sort the list in descending order using insertion sort
def insertion_sort_descending(integers_list):
    """
    Purpose: Sorts a list of integers in descending order using the
             insertion sort algorithm.

    Parameters:
    integers_list (list of int): The list of integers to be sorted

    Returns:
    list of int: The sorted list in descending order
    """

    # Start from the second element (index 1) as the first is assumed sorted
    for i in range(1, len(integers_list)):

        # The element we want to insert into the sorted part
        key = integers_list[i]

        # The index of the last element in the sorted part
        j = i - 1

        # Move elements of integers_list[0..i-1] that are smaller than key
        # to one position ahead of their current position (Shift to the right)
        while j >= 0 and key > integers_list[j]:
            integers_list[j + 1] = integers_list[j]
            j = j - 1

        # Insert the key into its correct position
        integers_list[j + 1] = key

    return integers_list


# =============================================================================
# PART (iii): RUNNING THE PROGRAM
# =============================================================================

# Define the filename
filename = 'input.txt'

# 1. Create the file manually
# We open in 'w' (write) mode to create the file and add the numbers
my_output_file = open(filename, 'w')
my_output_file.write("1 7 13 2 57 18 6 21 33 22 15")
my_output_file.close()

# 2. Call the function from Part (i) to read the data
# This function handles its own open/close internally
raw_data = read_integers_from_file(filename)

# 3. Call the function from Part (ii) to sort the data (The Extension)
sorted_data = insertion_sort_descending(raw_data)

# 4. Display the results
print("The integers sorted in descending order are:")
print(sorted_data)


```
